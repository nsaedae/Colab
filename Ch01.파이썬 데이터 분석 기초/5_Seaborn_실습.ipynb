```python
"""
날짜 : 2022/04/26
이름 : 김철학
내용 : 파이썬 Seaborn 실습하기

Seaborn
 - Seaborn 라이브러리는 matplotlib 라이브러리를 기반으로 만들어진 라이브러리
 - matplotlib 보다 코드가 쉽고 간결하고 고급 시각화를 지원
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# 막대그래프
plt.title('countplot')
sns.countplot(data=tips, x='day')
plt.show()
```

```python
#히스토그램
plt.title('histogram')
sns.distplot(tips['total_bill'])
plt.show()
```

```python
#산점도
plt.title('scatter')
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#박스상자
plt.title('boxplot')
sns.boxplot(data=tips, x='time', y='total_bill')
plt.show()
```

```python
# 쌍그래프 시각화
sns.pairplot(tips)
plt.show()
```
